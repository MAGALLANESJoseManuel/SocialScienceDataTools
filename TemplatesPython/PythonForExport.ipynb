```python
import networkx as nx
from itertools import combinations
from bs4 import BeautifulSoup as BS
import urllib2
import os
import matplotlib.pyplot as plot
import pandas as pd
%matplotlib inline
import re # library to use regular expressions in Python
```

We will create a function in Python for the cleaning:

```python
def clean_cell(dirtyCellAsString):
    cleanedCell=dirtyCellAsString.replace("million", "000000")
    cleanedCell=cleanedCell.replace(" ", "")
    cleanedCell=cleanedCell.replace(",", "")
    cleanedCell=re.sub(r'[0-9]-|[0-9][0-9]-|[0-9][0-9][0-9]-',"", cleanedCell)
    cleanedCell=re.sub(r'\[[\S ]]',"", cleanedCell)
    cleanedCell=re.findall(r'\d+',cleanedCell)  
    return cleanedCell
```

```python
#empty graph RE creation
countriesGraphFullClean=nx.Graph()
```

```python
os.chdir("/Users/josemagallanes/Documents/GITHUBrepositories/Tutorials/TemplatesPython") 
openHtml = open("warsList.html","r")  
soup = BS(openHtml,'lxml')
openHtml.close()

#getting data for nodes and edges
allTables = soup.findAll('table', {'class': 'sortable wikitable'})
for eachTable in allTables:
        allRows=eachTable.find_all("tr")
        for eachRow in range(1,len(allRows)):
            allCells = allRows[eachRow].find_all("td")
            
            #start - cleaning
            nameOfConflict=allCells[2].get_text().replace(u'\u2013', '-')            
            start=clean_cell(allCells[0].string)            
            end=clean_cell(allCells[1].string)                        
            fatalities=clean_cell(allCells[4].get_text())
            #end - cleaning
            
            #computing and exceptions
            try:
                duration=2+float(end[0]) - float(start[0])                
            except (ValueError,IndexError):
                duration=1
            try:
                fatalities=2+float(fatalities[0])
            except (ValueError,IndexError):
                fatalities=1
            
            #countries and coordinates from external links
            countries=[country.get_text() for country in allCells[3].find_all('a')]
            countries=['China' if country=="PRC" or country=="People's Republic of China" else country for country in countries]
            linksEachCountry=['http://en.wikipedia.org'+country.get('href') for country in allCells[3].find_all('a')] 
            coords=[]     
            for link in linksEachCountry:
                if link=='http://en.wikipedia.org/wiki/Soviet_Union':
                    link='http://en.wikipedia.org/wiki/Russia'
                subSoup = BS(urllib2.urlopen(link))
                lati,longi=subSoup.find_all("span",{'class': 'geo'})[0].string.split('; ')
                coords.append((float(lati),float(longi))) 
            countryInfo=zip(countries,coords)            
            
            #populating graph
            for country,(lati,longi)in countryInfo:
                countriesGraphFullClean.add_node(country, latitude=lati,longitude=longi)
            for pair in combinations (countries,2):
                countriesGraphFullClean.add_edge(*pair,name=nameOfConflict,duration=duration,fatalities=fatalities)
```

This network data will be saved. No more interaction with the web: 

```python
nx.write_graphml(countriesGraphFullClean, "WarsPython.graphml",encoding='utf-8')
```

We call it this way:

```python
countriesNet=nx.read_graphml("WarsPython.graphml")
```

```python
nx.draw(countriesNet)
```

But let's use the coordinates:

```python
X=nx.get_node_attributes(countriesNet,'longitude')
Y=nx.get_node_attributes(countriesNet,'latitude')
posNet=dict((country,(X[country],Y[country])) for country in X if country in Y)
nx.draw(countriesNet,pos=posNet)
```

R has many interesting features for Networ analysis, but more important, R has libraries for doing statictical models on networks, which are not available, to my knowledge, in Python. So if we need to use those libraries, we need to prepare a pair of files for R.  

First, the "edgelist" file (including the edge attributes only):

```python
import csv

with open('WarsForR.csv', 'w') as fileEdgeList:
    wr = csv.writer(fileEdgeList , delimiter=',')
    row="from","to","fatalities",'duration','name'
    wr.writerow(row)
    for c1,c2,c1c2Dict in countriesNet.edges(data=True):
        row=c1,c2,c1c2Dict['fatalities'],c1c2Dict['duration'],c1c2Dict['name']
        wr.writerow(row)
```

Finally, the attributes of the nodes will come in this file:

```python
with open('WarsForR_attr.csv', 'w') as fileNodeAttributes:
    wr = csv.writer(fileNodeAttributes, delimiter=',')
    row="country",'longitude','latitude'
    wr.writerow(row)
    for country,countryDict in countriesNet.nodes(data=True):
        row=country,countryDict['longitude'],countryDict['latitude']
        wr.writerow(row)
```
